```python
import math

def arc_counter( formats, counts):
    arcs = 0
    for f, c in zip(formats, counts):
        arcs = max(arcs, c/f)
    return arcs
```

```python
formats = [16,8]
counts = [400*50, 120*50]
```

```python
arc_counter(formats, counts)
```

```python
def full_arc(a, b, a_mul, b_mul, total):
    return a * a_mul + b * b_mul == total
```

```python
product = itertools.product(a7,a6)
goodones = [(a,b) for a, b in product if full_arc(a, b, 1, 2, 36)]
print(goodones)
```

```python
minarc = 10000
combotext=""
    
for combo in goodones:
    if minarc > arc_counter(combo,counts):
        minarc=arc_counter(combo,counts)
        combotext="na arc{} : arc{}".format(combo, minarc)
print(combotext)
```

```python
arc_counter(goodones[0],counts)
```
